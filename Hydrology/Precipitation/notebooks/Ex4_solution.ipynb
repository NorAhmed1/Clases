### Derive the hypsometric curve

```python
# define elevation thresholds 
Zs = np.arange(start=0, stop=1701, step=100)
Zs
```

```python
# n. of cells in the catchment
ncells = np.sum(~np.isnan(dem))
ncells
```

Example: 100 m elevation

```python
# set a threshold
Z = 100
```

```python
# n. of cells below the threshold
n = np.sum(dem < Z)
n
```

```python
# value in the hypsometric 
n / ncells
```

Loop for all elevation thresholds

```python
# table where to save the data
hypso = pd.Series(index=Zs)
hypso.head()
```

```python
# compute the hypsometric curve
for Z in Zs:
    hypso[Z] = np.sum(dem < Z) / ncells
hypso.head()
```

```python
# plot the hypsometric curve
plt.plot(Zs, hypso)
plt.title('Hypsometric curve')
plt.xlabel('elevation (masl)')
plt.ylabel('area (-)');
```

### Calculate the weights

```python
Zbands = []
weights = []
for Zi, Zj in zip(Zs[:-1], Zs[1:]):
    Zbands.append(np.mean([Zi, Zj]))
    weights.append(hypso[Zj] - hypso[Zi])
```

```python
weights = pd.Series(data=weights, index=Zbands)
weights
```

```python
plt.plot(weights)
plt.xlabel('elevation (masl)')
plt.ylabel('weight (-)');
```
