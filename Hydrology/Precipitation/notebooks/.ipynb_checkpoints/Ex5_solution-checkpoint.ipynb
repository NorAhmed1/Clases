```python
import numpy as np

import pandas as pd
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
sns.set_context('notebook')
```

## Import data

```python
pcp_h = pd.read_csv('../data/hourly_precipitation_Oviedo_AEMET+SAIH.csv', parse_dates=True, index_col='datetime')
pcp_h.head()
```

```python
# plot original data
plt.figure(figsize=(15, 3.5))
plt.plot(pcp_h.index, pcp_h, linewidth=.2)
plt.ylabel('P (mm)');
```

## Generate series of annual maxima
### Example for 2h storm duration

```python
# series of 2h rainfall intensity
pcp_2h = pcp_h.rolling(2, center=True).sum()
int_2h = pcp_2h / 2
int_2h.columns = ['I[mm/h]']
int_2h.head()
```

```python
# series of annual maximum intensity
annualMax_2h = int_2h.groupby(by=int_2h.index.year).max()
annualMax_2h.head()
```

```python
# visualize the data
sns.distplot(annualMax_2h, bins=8);
```

### loop for storn durations

```python
# durations to study
D = np.array([1, 2, 4, 8, 16, 24])
```

```python
# series of annual maximum intensity
annualMax = pd.DataFrame(index=annualMax_2h.index)

for d in D:
    int_d = pcp_h.rolling(d, center=True).sum() / d
    annualMax[d] = int_d.groupby(by=int_d.index.year).max()
annualMax
```

### Fit a GEV distribution to the series of annual maxima

```python
from scipy.stats import genextreme
from statsmodels.distributions.empirical_distribution import ECDF
```

#### Example: 2-hour storm

```python
# visualize data
sns.distplot(annualMax[2], bins=8, fit=genextreme);
```

```python
# fit a GEV
parGEV_2h = genextreme.fit(annualMax[2], f0=0)
parGEV_2h
```

```python
# fit the empirical distribution
ecdf_2h = ECDF(annualMax[2])
```

```python
annualMax[2].shape
```

```python
# plot
I = np.linspace(start=0, stop=25, num=100)
plt.plot(I, genextreme(*parGEV_2h).cdf(I))
plt.plot(annualMax[2], ecdf_2h(annualMax[2]), '.');
```

#### Loop for all storm durations

```python
# fit paramaters for each durations
parameters = pd.DataFrame(index=['xi', 'mu', 'sigma'], columns=D)
for d in D:
    parameters[d] = genextreme.fit(annualMax[d], f0=0)
parameters
```

```python
# plot
I = np.linspace(start=0, stop=35, num=100)
for d in D:
    plt.plot(I, genextreme(*parameters[d]).cdf(I))
    plt.scatter(annualMax[d], ECDF(annualMax[d])(annualMax[d]), s=5, label=str(d) + ' h')
plt.legend();
```

### Analytical IDF

```python
def IDF_type_I(x, a, b, c, d):
    """Estimate precipitation intensity given a return period and a duration
    
    I = (a * R + b) / (D + c)**d
    
    Inputs:
    -------
    x:     list [2, 1]. Values of return period and duration
    a, b, c, d: float.Parameters of the IDF curve
    
    Output:
    -------
    I:     float. Precipitation intensity (mm/h)
    """
    
    I = (a * x[0] + b) / (x[1] + c)**d
    
    return I
```

```python
def IDF_type_II(x, a, b, c, d):
    """Estimate precipitation intensity fiven a return period and a storm duration using the analytical IDF curve type II:
    
    I = (a * R + b) / (D**c + d)   
    
    Input:
    ------
    x:         list [2x1]. Values of return period (years) and duration (h)
    a:         float. Parameter of the IDF curve
    b:         float. Parameter of the IDF curve
    c:         float. Parameter of the IDF curve
    d:         float. Parameter of the IDF curve
    
    Output:
    -------
    I:         float. Precipitation intensity (mm/h)"""
    
    I = (a * x[0] + b) / (x[1]**c + d)
    
    return I

def IDF_type_III(x, a, b, c, d):
    """Estimate precipitation intensity fiven a return period and a storm duration using the analytical IDF curve type III:
    
    I = a * R**b / (D + c)**d
    
    Input:
    ------
    x:         list [2x1]. Values of return period (years) and duration (h)
    a:         float. Parameter of the IDF curve
    b:         float. Parameter of the IDF curve
    c:         float. Parameter of the IDF curve
    d:         float. Parameter of the IDF curve
    
    Output:
    -------
    I:         float. Precipitation intensity (mm/h)"""
    
    I = a * x[0]**b  / (x[1] + c)**d
    
    return I

def IDF_type_IV(x, a, b, c, d):
    """Estimate precipitation intensity fiven a return period and a storm duration using the analytical IDF curve type IV:
    
    I = a * R**b / (D**c + d).    
    
    Input:
    ------
    x:         list [2x1]. Values of return period (years) and duration (h)
    a:         float. Parameter of the IDF curve
    b:         float. Parameter of the IDF curve
    c:         float. Parameter of the IDF curve
    d:         float. Parameter of the IDF curve
    
    Output:
    -------
    I:         float. Precipitation intensity (mm/h)"""
    
    I = (a * x[0]**b) / (x[1]**c + d)
    
    return I 
```

#### Fit the analytical IDF

```python
from scipy.optimize import curve_fit
```

##### Dependent variable in the IDF curve: intensity

```python
IDFe
```

```python
I = np.empty((0))
for d in D:
    I = np.concatenate((I, IDFe[d]))
I.shape
```

```python
I
```

##### Independent variable in the IDF curve: paired values of return period and duration

```python
R, D
```

```python
RR, DD = np.meshgrid(R, D)
RR.shape, DD.shape
```

```python
RR = RR.flatten()
DD = DD.flatten()
```

```python
RR, DD
```

```python
RD = np.vstack([RR, DD])
RD.shape
```

```python
I
```

```python
RD
```

```python
IDFe
```

##### FIt the curve

```python
# set the type of IDF
curve = IDF_type_IV
```

```python
parIDF = curve_fit(curve, RD, I)[0]
parIDF
```

```python
# save optimized parameters
IDFa = pd.DataFrame(data=parIDF, index=['a', 'b', 'c', 'd']).transpose()
IDFa
```

```python
IDFa.to_csv('../output/Ex5_Parameters analytical IDF.csv', index=False, float_format='%.5f')
```

##### Visualize the fit

```python
fig = plt.figure(figsize=(10, 4))
plt.xlim(0, 25)
plt.xlabel('duration (h)')
plt.ylabel('intensity (mm/h)')
color = ['tan', 'darkkhaki', 'olive']

D_ = np.linspace(1, 25, 100)

for i, Tr in enumerate(R):
    R_ = np.ones_like(D_) * Tr
    I_ = curve((R_, D_), *parIDF)
    
    plt.scatter(D, IDFe.loc[Tr, :], color=color[i], s=10, label='Tr = ' + str(int(Tr)) + ' years')
    plt.plot(D_, I_, color=color[i])
    
fig.legend(loc=8, ncol=3);

plt.savefig('../output/Ex5_analytical IDFs.png', dpi=300)
```
